### 2번 데이터프레임에 대해 전처리

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
# 조사, 어미, 구두점 등 제거
def okt_clean(text):
    clean_text = []
    for word in okt.pos(text, stem=True):
        if word[1] not in ['Josa', 'Eomi', 'Punctuation']:
            clean_text.append(word[0])

    return " ".join(clean_text)
```

```python
df1.loc[:,'okt_symptoms']= df1['symptoms'].apply(okt_clean)
df1['okt_symptoms'][:2]
```

### Tokenizing

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
documents = [TaggedDocument((disease + ' ' + symptoms).split(), [str(i)])
             for i, (disease, symptoms) in zip(df1.index, zip(df1['disease_korean_title'], df1['okt_symptoms']))]

# Doc2Vec 모델을 학습시킵니다.
model = Doc2Vec(documents, vector_size=100, window=5, min_count=1, workers=4)
```

```python
df1.info()
```

```python
def predict_diseases(model, symptom, threshold=0.95, topn=20):
    symptom_words = symptom.split()
    symptom_vector = model.infer_vector(symptom_words)
    similar_diseases = model.dv.most_similar([symptom_vector], topn=topn)
    return list(dict.fromkeys([df1.at[int(i), 'disease_korean_title'] for i, similarity in similar_diseases if similarity >= threshold]))
symptom = input('증상을 입력하세요: ')
diseases = predict_diseases(model, symptom)
print('해당 증상과 가장 유사한 질병들은 다음과 같습니다:')
for disease in diseases:
    print(disease)
```
